```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the CSV file
data2 = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='ISO-8859-1')

data2 = data2.drop(columns=['Product Description'])

# Select numeric columns for X
X = data2.select_dtypes(include=[np.number])

# Extract the target variable (y)
y = data2['Sales']

# Before applying PCA, print the correlation matrix
initial_correlation_matrix = X.corr()
print("Initial Correlation Matrix:")
print(initial_correlation_matrix['Sales'].sort_values(ascending=False))

# Impute missing values (replace NaN with mean)
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Determine the number of components to keep
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
num_components_to_keep = np.argmax(cumulative_variance_ratio >= 0.95) + 1

# Reduce the features using the optimal number of components
X_final = X_pca[:, :num_components_to_keep]

# Display the removed columns
columns_before_pca = X.columns.tolist()
columns_after_pca = [f'PC{i+1}' for i in range(num_components_to_keep)]
columns_removed = set(columns_before_pca) - set(columns_after_pca)

print("Columns Removed by PCA:", columns_removed)

# Convert the set to a list
columns_removed_list = list(columns_removed)

# Create a DataFrame with the removed columns
removed_columns_df = X[columns_removed_list].copy()

# Concatenate the removed columns DataFrame with the 'Sales' column
removed_columns_df['Sales'] = y

# Display the correlation values as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(removed_columns_df.corr()[['Sales']].sort_values(by='Sales', ascending=False),
            annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation between Removed Columns and Sales')
plt.show()
```
